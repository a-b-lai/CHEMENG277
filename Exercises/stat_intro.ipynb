Introduction to Frequentist Statistics
=============================

In this Jupyter Notebook...

* Analyze the distribution of measurements of pesticide content in water supplies in California (measurements between 6/1/2020 and 8/1/2020, provided by EPA)

* Determine moments of the distribution based on an ensemble average and using the distribution

* Demonstrate how a predicted distribution based on the first two moments can lead to erroneous behavior

* Note other measured properties to explore include Nitrate, Nitrite, and Arsenic (see data file for more properties)

Import the necessary modules

```python
import numpy as np
import sys
import pandas as pd
sys.path.append('../')
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/ca_water.csv", low_memory = False)
exp_data = data[['CharacteristicName', 'ResultMeasureValue']]
```

```python
def isfloat(value):
    try:
        float(value)
        return True
    except ValueError:
        return False
```

Select the data from the full dataset with a 'CharacteristicName' given by 'Orthophosphate'

```python
x = np.array(exp_data[exp_data['CharacteristicName'] == 'Orthophosphate']['ResultMeasureValue'])

pesticide = np.array([])
for i in range(len(x)):
    if isfloat(x[i]):
        if np.logical_not(np.isnan(float(x[i]))):
            pesticide = np.append(pesticide, float(x[i]))
    else:
        pesticide = np.append(pesticide, 0)
```

Generate a probability density function from the histogram of the pesticide concentrations

```python
x0 = 0
xf = 5
n_bin = 400
x = np.linspace(x0, xf, n_bin)

delta_x = x[1] - x[0]
x_bin = np.append(x,x[-1] + delta_x) - delta_x / 2
prob, x_bin = np.histogram(pesticide, bins=x_bin, density = True)

plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

plt.plot(x, prob)

plt.xlabel('Concentration of orthophosphate pesticides [$mg/l$]')
plt.ylabel('Probability density')

plt.show()
```

Calculate the first two moments directly from the data

```python
conc_ave = np.mean(pesticide) #based on lecture notes, we can fit an exponential function with just the first two moments
conc2_ave = np.mean(pesticide ** 2)
```

```python
print(conc_ave)
print(conc2_ave)
```

Calculate the first two moments from the histogram

```python
conc_ave_hist = np.sum(x ** 1 * prob * delta_x) #x is a vector of bins 
conc2_ave_hist = np.sum(x ** 2 * prob * delta_x)
```

```python
print(conc_ave_hist)
print(conc2_ave_hist)
```

Assume a model for the probability density function for concentration $x$, given by $f_{x}(x) = \lambda \exp (-\lambda x)$. Plot this model based on the property $\langle x \rangle = 1/\lambda$

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

lam = 1 / conc_ave_hist

plt.plot(x, prob)
plt.plot(x, lam * np.exp(-lam * x))

plt.xlabel('Concentration of orthophosphate pesticides [$mg/l$]')
plt.ylabel('Probability density')

plt.show()
```

Now plot the first few moments of the observed distribution versus the predictions from our model

```python
n_m = 6
moment = np.zeros((n_m, 2))
moment_th = np.zeros(n_m)

lam = 1 / conc_ave_hist
for i_m in range(n_m):
    moment[i_m, 0] = np.sum(x ** i_m * prob * delta_x)
    moment[i_m, 1] = np.mean(pesticide ** i_m)
    moment_th[i_m] = np.math.factorial(i_m) / lam ** i_m
```

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

lam = 1 / conc_ave_hist
mom_ind = np.linspace(0, n_m - 1, n_m)

plt.scatter(mom_ind, moment[:,0], color = 'C0', s = 50)
plt.scatter(mom_ind, moment[:,1], color = 'C1', s = 50)
plt.scatter(mom_ind, moment_th, color = 'C2', s = 50)

plt.xlabel('m')
plt.ylabel(r'Value of the mth moment $\langle x^{m} \rangle$')

plt.show()
```

Plot the probability density function on a logarithmic scale to show the tail of the distribution

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

lam = 1 / conc_ave_hist

plt.scatter(x, prob)
plt.plot(x, lam * np.exp(-lam * x))

#plt.xscale('log')
plt.yscale('log')

plt.xlabel('Concentration of orthophosphate pesticides [$mg/l$]')
plt.ylabel('Probability density')

plt.ylim([1e-2, 10])
plt.show()
```
