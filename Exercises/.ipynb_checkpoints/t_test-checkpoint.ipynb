# Global Temperature Anomalies

Global temperature anomalies indicate deviations from long-term average temperatures, with positive values suggestive of global warming.
The data provides average annual temperature anomalies between 1900 and 2015 for three latitude bands: 90N-23.6N ('North'), 23.6N-23.6S ('Central'), 23.6S-90S ('South').
We will begin using t-tests to demonstrate that temperature anomalies are significantly higher during more recent years. 
We will then use t-tests to check whether global warming disproportionately affects one of the latitude bands.

This exercise demonstrates:

    1. Assessing assumptions of the t-test
    2. Applying t-tests to compare means between two populations

Data source: https://cdiac.ess-dive.lbl.gov/ftp/trends/temp/hansen/norlowsou.txt

## Load Data

```python
import pandas as pd
import numpy as np

temp_anomalies = pd.read_csv("data/temp_anomalies_lat_bands.csv", header=0, delimiter='\t')
temp_anomalies.head()
```

## Visualize Data

```python
import matplotlib.pyplot as plt

plt.scatter(temp_anomalies.Year, temp_anomalies.North, label="90 N - 23.6 N")
plt.scatter(temp_anomalies.Year, temp_anomalies.Central, label="23.6 N - 23.6 S")
plt.scatter(temp_anomalies.Year, temp_anomalies.South, label="23.6 S - 90 S")
plt.title("Annual Temperature Anomalies")
plt.xlabel("Year")
plt.ylabel("Temperature Anomaly (C)")
plt.legend()
plt.show()
```

## Subset Data for Comparisons

Isolate points from the last 20 years in the data set and compare to the distribution of all previous years using a t-test.

```python
# Split temperature anomalies between most recent 10 years and remaining history
history = temp_anomalies.iloc[:-20,:]
recent = temp_anomalies.iloc[-20:,:]

# Compare the distributions of temperature anomalies between time frames
bins = np.linspace(-0.5, 1, 20)
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(5, 9), sharex=True, sharey=True)
ax0, ax1, ax2 = axes.flatten()
ax0.hist(recent.North, label="Recent", bins=bins, alpha=0.5)
ax0.hist(history.North, label="Historic", bins=bins, alpha=0.5)
ax0.legend()
ax1.hist(recent.Central, label="Recent", bins=bins, alpha=0.5)
ax1.hist(history.Central, label="Historic", bins=bins, alpha=0.5)
ax2.hist(recent.South, label="Recent", bins=bins, alpha=0.5)
ax2.hist(history.South, label="Historic", bins=bins, alpha=0.5)
fig.text(0.5, -0.01, 'Temperature Anomaly', ha='center')
fig.text(-0.01, 0.5, 'Frequency', va='center', rotation='vertical')
plt.tight_layout()
plt.show()
```

## Assess Normality

The normality assumption is required for comparison with t-tests. 
Use a Quantile-Quantile plot to assess normality.

```python
 import scipy.stats as stats

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(9, 6), sharex=True, sharey=True)
ax0, ax1, ax2, ax3, ax4, ax5 = axes.flatten()

stats.probplot(recent.North, dist="norm", plot=ax0)
ax0.set_title("North, Recent")
ax0.set_xlabel("")
stats.probplot(recent.Central, dist="norm", plot=ax1)
ax1.set_title("Central, Recent")
ax1.set_xlabel("")
ax1.set_ylabel("")
stats.probplot(recent.South, dist="norm", plot=ax2)
ax2.set_title("South, Recent")
ax2.set_xlabel("")
ax2.set_ylabel("")

stats.probplot(history.North, dist="norm", plot=ax3)
ax3.set_title("North, Historic")
stats.probplot(history.Central, dist="norm", plot=ax4)
ax4.set_title("Central, Historic")
ax4.set_ylabel("")
stats.probplot(history.South, dist="norm", plot=ax5)
ax5.set_title("South, Historic")
ax5.set_ylabel("")

plt.subplots_adjust(top = 0.99, bottom=0.01, hspace=0.2, wspace=0.3)
plt.show()
```

The data appears reasonably normally distributed. Use t-tests to compare the distributions of temperature anomalies between the most recent 10 years and the remainder of history.

**Null Hypothesis:** The mean temperature anomaly is the same for both subsets.

**Alternative Hypothesis:** The mean temperature anomaly is greater for the subset of recent years.

**Significance Threshold:** 0.05

```python
def Welch_Satterthwaite_equation(stdev_1, stdev_2, n_1, n_2):
    """Get the degrees of freedom for a two-sample t-test with unqual sample variances.

    When two samples have an unequal variance, the "approximate effective degrees of
    freedom" for a t-test comparing their means is given by the Welch-Satterthwaite
    equation. The origin of this approximation is beyond the scope of this class, and is
    discussed in:

    1. https://doi.org/10.2307/3002019 (Satterthwaite 1946)
    2. https://doi.org/10.2307/2332510 (Welch 1947)

    Parameters
    ----------
    stdev_1, stdev_2 : float
        Standard deviation of the first and second samples
    n_1, n_2 : float
        Number of data points in the first and second samples

    Returns
    -------
    float
        Degrees of freedom for a two-sample
    """
    return round(
        (stdev_1**2 / n_1 + stdev_2**2 / n_2) ** 2 /
            (stdev_1 ** 4 / (n_1 ** 2 * (n_1 - 1)) +
                stdev_2 ** 4 / (n_2 ** 2 * (n_2 - 1)))
    )


def t_test_one_sided(set_1, set_2):
    """Run a t-test on the two data sets.

    Parameters
    ----------
    set_1: np.ndarray
        First sample of data points in hypothesis test
    set_2: np.ndarray
        Second sample of data points in hypothesis test

    Returns
    -------
    P-value for a one sided hypothesis test where the alternative hypothesis is
        mean(set_1) > mean(set_2)
    """
    # Calculate means, standard deviations, and sample sizes
    mean_1 = np.average(set_1)
    mean_2 = np.average(set_2)
    stdev_1 = np.std(set_1)
    stdev_2 = np.std(set_2)
    n_1 = len(set_1)
    n_2 = len(set_2)

    # Approximate degrees of freedom using the Welch-Satterthwaite equation
    df = Welch_Satterthwaite_equation(stdev_1, stdev_2, n_1, n_2)

    # Generate t-statistic
    t = (mean_1 - mean_2) / (np.sqrt(stdev_1**2/n_1 + stdev_2**2/n_2))
    
    # return p-value
    return 1 - stats.t.cdf(t, df=df)
    

p_val_north = t_test_one_sided(recent.North, history.North)
print("P-value, Northern Latitude Band: " + str(p_val_north))
print()
p_val_central = t_test_one_sided(recent.Central, history.Central)
print("P-value, Central Latitude Band: " + str(p_val_central))
print()
p_val_south = t_test_one_sided(recent.South, history.South)
print("P-value, Central Latitude Band: " + str(p_val_south))
```

With p-values below our significance threshold, there is sufficient evidence to reject the null hypothesis and claim that temperature temperature anomalies are greatest during the recent years of the data set.

## Compare Latitude Bands

```python
North_Central = temp_anomalies.North - temp_anomalies.Central
Central_South = temp_anomalies.Central - temp_anomalies.South
North_South = temp_anomalies.North - temp_anomalies.South

# Plot histograms of pairwise differences in temperature anomalies between latitude bands
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(9, 3), sharex=True, sharey=True)
ax0, ax1, ax2 = axes.flatten()
ax0.hist(North_Central)
ax0.set_title("North - Central")
ax1.hist(Central_South)
ax1.set_title("Central - South")
ax2.hist(North_South)
ax2.set_title("North - South")
fig.text(0.5, -0.01, 'Pairwise Difference in Temperature Anomaly', ha='center')
fig.text(-0.01, 0.5, 'Frequency', va='center', rotation='vertical')
plt.tight_layout()
plt.show()

# Generate quantile-quantile plots assessing normality in pairwise differences
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(9, 3), sharex=True, sharey=True)
ax0, ax1, ax2 = axes.flatten()
stats.probplot(North_Central, dist="norm", plot=ax0)
ax0.set_title("North - Central")
stats.probplot(Central_South, dist="norm", plot=ax1)
ax1.set_title("Central - South")
ax1.set_ylabel("")
stats.probplot(North_South, dist="norm", plot=ax2)
ax2.set_title("North - South")
ax2.set_ylabel("")
plt.tight_layout()
plt.show()
```

The distribution in pairwise differences in temperature anomalies between latitude bands appear normally distributed.
Use t-tests to check if there is a statistically significant difference in temperature anomaly between each pair of latitude bands.

**Null Hypothesis:** The mean difference in temperature anomaly between latitude bands is zero.

**Alternative Hypothesis:** The mean difference in temperature anomaly between latitude bands is not zero.

**Significance Threshold:** 0.05

```python
def t_test_one_sample(set_1):
    """Run a one-sample, two sided t-test on a dataset.

    The null hypothesis is that the mean is zero.
    The alternative hypothesis is that the mean is not zero.

    Parameters
    ----------
    set_1 : np.ndarray
        Data set on which the mean is being compared
    
    Returns
    -------
    P-value for one sample, two-sided t-test where the null hypothesis is that the
        mean is zero and the alternative hypothesis is that the mean is not zero.
    """
    # Calculate mean, standard deviation, sample size, degrees freedom
    mean_1 = np.average(set_1)
    stdev_1 = np.std(set_1)
    n_1 = len(set_1)
    df = n_1 - 1
    
    # Calculate t-statistic
    t = mean_1 / (stdev_1 / np.sqrt(n_1))

    # return p-value
    return stats.t.sf(np.abs(t), df) * 2    # stats.t.sf is the "survival function" = 1 - cdf

p_val_north_central = t_test_one_sample(North_Central)
print("P-value, North-Center=0: " + str(p_val_north_central))
print()
p_val_central_south = t_test_one_sample(Central_South)
print("P-value, Center-South=0: " + str(p_val_central_south))
print()
p_val_north_south = t_test_one_sample(North_South)
print("P-value, North-South=0: " + str(p_val_north_south))
print()
```

With p-values below the significance threshold of 0.05 for all pairs, there is sufficient evidence to conclude that the mean temperature anomalies differ between latitude bands.
