Regularization Methods: Ridge versus LASSO
====================================

In this discussion...

* Apply Ridge and LASSO regularization methods for noisy data

* Show the impact of signal to noise on the fit parameters for varying $\lambda$

* Demonstrate the impact of the contributions to the cost function for the two methods


Import the necessary modules

```python
import numpy as np
import sys
import pandas as pd
import scipy as sp
from sklearn import linear_model
sys.path.append('../')
import matplotlib.pyplot as plt
```

Define the model as a linear model with two covariates with Gaussian random noise added to the data

```python
# Specify linear model parameters (y = a1 * t1 + a2 * t2)
a1 = 2
a2 = 8
sigma = 4
```

```python
n_e = 200   # Number of data points

t1 = np.zeros(n_e)
t2 = np.zeros(n_e)

for i in range(n_e):
    t1[i] = np.random.normal()      # Generate normally distributed random variable
    t2[i] = np.random.normal()
```

Center and standardize the data

```python
# Center and standardize the covariates
t1 = t1 - np.mean(t1)
t1 = t1 / np.sqrt(np.sum(t1 ** 2) / (n_e - 1))
t2 = t2 - np.mean(t2)
t2 = t2 / np.sqrt(np.sum(t2 ** 2) / (n_e - 1))

plt.figure()
bins = np.linspace(-4, 4, 16)
plt.hist(t1, color='red', alpha=0.5, label="$t_1$", bins=bins)
plt.hist(t2, color='blue', alpha=0.5, label="$t_2$", bins=bins)
plt.legend(loc="best")
plt.show()
```

```python
# Apply the linear model
x_cen = a1 * t1 + a2 * t2 + sigma * np.random.normal(size = n_e)    # Simulate response with Gaussian distributed noise
x_fit = a1 * t1 + a2 * t2                                           # Generate predicted response
```

Plot the data 

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

plt.scatter(t1, x_cen, s = 30)  # Plot the outcome against each covariate
plt.scatter(t2, x_cen, s = 30)

plt.ylabel('Outcome $x$')
plt.xlabel(r'Covariate $\Delta \hat{t}^{(1)}$ (blue) and $\Delta \hat{t}^{(2)}$ (orange)')

plt.show()
```

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

plt.scatter(x_fit, x_cen, s = 30)       # Plot the relationship between the model fit and the observed outcome
plt.plot(np.linspace(-20, 20, 10), np.linspace(-20, 20, 10), color = 'C1')


plt.xlabel('Noiseless outcome $x$')
plt.ylabel('Outcome $x$')

plt.show()
```

Perform a multivariable linear regression of the data (i.e. without regularization)

```python
# Fit the model without regularization
n_t = 2
mt = np.zeros(n_t+1)
tt = np.zeros((n_t+1, n_t+1))

t_cen = np.transpose(np.array([t1, t2]))

mt[0] = 0
tt[0,0] = 1
for i in range(n_t):
    mt[i+1] = np.mean(x_cen * t_cen[:,i])
    tt[i+1,0] = np.mean(t_cen[:,i])
    tt[0,i+1] = np.mean(t_cen[:,i])
    for j in range(n_t):
        tt[i+1,j+1] = np.mean(t_cen[:,i] * t_cen[:,j])
```

```python
tt_inv = np.linalg.inv(tt)
a = np.dot(tt_inv, mt)
```

```python
a   # Notice that the intercept term is nearly zero
```

Perform Ridge regression for various values of $\lambda$

```python
lam_vec = np.logspace(-3,2,11)      # Specify vector of 11 lambdas ranging from 0.001 to 100
a_ridge = np.zeros((len(lam_vec), n_t+1))

ip_mat = np.eye(n_t + 1)
ip_mat[0,0] = 0
for i in range(len(lam_vec)):       # For each value of lambda, calculate the coefficients by ridge regression
    tt_inv = np.linalg.inv(tt + lam_vec[i] * ip_mat)
    a_ridge[i,:] = np.dot(tt_inv, mt)
```

```python
a_ridge
```

Perform LASSO regression for various values of $\lambda$

```python
lam_vec = np.logspace(-3,2,11)
a_lasso = np.zeros((len(lam_vec), n_t+1))

ip_mat = np.eye(n_t + 1)
ip_mat[0,0] = 0
for i in range(len(lam_vec)):       # For each value of lambda, calculate the coefficients by LASSO regression
    clf = linear_model.Lasso(alpha=lam_vec[i]/2, fit_intercept = False)
    clf.fit(t_cen, x_cen) 
    a_lasso[i,1:] = clf.coef_
```

```python
a_lasso
# Notice, LASSO is able to return coefficients of zero, thereby canceling the contribution of weakly correlated covariates
# As you increase the value of lambda, the model becomes less tolerant of noise
```

Compare the coefficients from Ridge and LASSO

```python
# Here we plot the magnitude of the each coefficient against values of lambda
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)


plt.scatter(lam_vec, np.abs(a_ridge[:,1]), color = 'C0')
plt.plot(lam_vec, np.abs(a_ridge[:,1]), color = 'C0', label="a1 - ridge")
plt.scatter(lam_vec, np.abs(a_lasso[:,1]), color = 'C0')
plt.plot(lam_vec, np.abs(a_lasso[:,1]),'--', color = 'C0', label="a1 - lasso")

plt.scatter(lam_vec, np.abs(a_ridge[:,2]), color = 'C1')
plt.plot(lam_vec, np.abs(a_ridge[:,2]), color = 'C1', label="a2 - ridge")
plt.scatter(lam_vec, np.abs(a_lasso[:,2]), color = 'C1')
plt.plot(lam_vec, np.abs(a_lasso[:,2]),'--', color = 'C1', label="a2 - lasso")


plt.xlabel('$\lambda$')
plt.ylabel('Magnitude of coefficient $|a_{k}|$')

plt.xscale('log')
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left")
#plt.savefig('figures/regularization/coef_ridge.pdf', bbox_inches='tight')
plt.show()
```

```python
# Here we plot the ratio of coefficients generated by Ridge and Lasso
# A value of 1 means the ratio of coefficients obtained by regression matches that of the model
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)


plt.scatter(lam_vec, (a2 / a1) * np.abs(a_ridge[:,1]) / np.abs(a_ridge[:,2]), color = 'C0')
plt.plot(lam_vec, (a2 / a1) * np.abs(a_ridge[:,1]) / np.abs(a_ridge[:,2]), color = 'C0', label="Ridge")

plt.scatter(lam_vec, (a2 / a1) * np.abs(a_lasso[:,1]) / np.abs(a_lasso[:,2]), color = 'C1')
plt.plot(lam_vec, (a2 / a1) * np.abs(a_lasso[:,1]) / np.abs(a_lasso[:,2]), color = 'C1', label="Lasso")

plt.xlabel('$\lambda$')
plt.ylabel('Ratio of coefficients $(|a_{1}| a_{2_{0}})/ (|a_{2}|  a_{1_{0}})$')
plt.legend(loc='best')

plt.xscale('log')
#plt.savefig('figures/regularization/coef_ridge.pdf', bbox_inches='tight')
plt.show()
```

Plot the contributions to the cost function for varying $a_{1}$ and $a_{2}$ for $\lambda = 1$

```python
# Calculate cost function for various a2 values, holding a1 constant
n_a = 100
lam_ind = 6
a2_vec = np.linspace(0, 16, n_a)    # Generate the cost function for various a2 coefficients
j_0 = np.zeros(n_a)

for i in range(n_a):
    j_0[i] = np.sum((x_cen - a1 * t1 - a2_vec[i] * t2) ** 2)

j_reg_ridge = lam_vec[lam_ind] * n_e * (a1 ** 2 + a2_vec ** 2)          # Ridge regularization cost function
j_reg_lasso = lam_vec[lam_ind] * n_e * (np.abs(a1) + np.abs(a2_vec))    # LASSO regularization cosft function
```

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)


plt.plot(a2_vec, j_0, color = 'C0', label="SSE")
plt.plot(a2_vec, j_reg_ridge, color = 'C1', label="RIDGE adjust")
plt.plot(a2_vec, j_reg_lasso, '--', color = 'C1', label="LASSO adjust")

plt.plot(a2_vec, j_0 + j_reg_ridge, color = 'C2', label="RIDGE overall")
plt.plot(a2_vec, j_0 + j_reg_lasso, '--', color = 'C2', label="LASSO overall")

plt.ylim([0, 40000])

plt.xlabel('Value of the fit coefficient $a_{2}$')
plt.ylabel('Cost function contributions')
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left")
#plt.savefig('figures/regularization/coef_ridge.pdf', bbox_inches='tight')
plt.show()
```

```python
# Calculate cost function for various a1 values, holding a2 constant
n_a = 100
lam_ind = 6
a1_vec = np.linspace(0, 16, n_a)
j_0 = np.zeros(n_a)

for i in range(n_a):
    j_0[i] = np.sum((x_cen - a1_vec[i] * t1 - a2 * t2) ** 2)

j_reg_ridge = lam_vec[lam_ind] * n_e * (a1_vec ** 2 + a2 ** 2)
j_reg_lasso = lam_vec[lam_ind] * n_e * (np.abs(a1_vec) + np.abs(a2))
```

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)


plt.plot(a1_vec, j_0, color = 'C0', label="SSE")
plt.plot(a1_vec, j_reg_ridge, color = 'C1', label="RIDGE adjust")
plt.plot(a1_vec, j_reg_lasso, '--', color = 'C1', label="LASSO adjust")

plt.plot(a1_vec, j_0 + j_reg_ridge, color = 'C2', label="RIDGE overall")
plt.plot(a1_vec, j_0 + j_reg_lasso, '--', color = 'C2', label="LASSO overall")

plt.ylim([0, 40000])

plt.xlabel('Value of the fit coefficient $a_{1}$')
plt.ylabel('Cost function contributions')
plt.legend(bbox_to_anchor=(1.04,0.5), loc="center left")
#plt.savefig('figures/regularization/coef_ridge.pdf', bbox_inches='tight')
plt.show()
```
