Non-dimensionalization
===================

In this Jupyter Notebook...

* Analyze the penetration of chemotherapeutics into a solid tumor spheroid (see Thurber, Greg M., and K. Dane Wittrup. "Quantitative spatiotemporal analysis of antibody fragment diffusion and endocytic consumption in tumor spheroids." Cancer research 68, no. 9 (2008): 3334-3341.)

* Plot dimensional data for the predicted behavior (simulated) for three drugs with well-characterized properties

* Reframe the data using our dimensionless groups identified in lecture

Import the necessary modules

```python
import numpy as np
import sys
import pandas as pd
import scipy as sp
sys.path.append('../')
import matplotlib.pyplot as plt
```

Load the data and define the diffusion coefficient and rate of cell uptake for the following three drugs:

* Drug A - Doxorubicin 
        
    * Weinberg, Brent D., Ravi B. Patel, Agata A. Exner, Gerald M. Saidel, and Jinming Gao. "Modeling doxorubicin transport to improve intratumoral drug delivery to RF ablated tumors." Journal of controlled release 124, no. 1-2 (2007): 11-19.

    * Qian F, Stowe N, Liu EH, Saidel GM, Gao J. Quantification of in vivo doxorubicin transport from PLGA millirods in thermoablated rat livers. J Control Release. 2003;91(1–2):157–166.
    
* Drug B - Cisplatin

    * Modok, S., R. Scott, R. A. Alderden, M. D. Hall, H. R. Mellor, S. Bohic, Tina Roose, T. W. Hambley, and R. Callaghan. "Transport kinetics of four-and six-coordinate platinum compounds in the multicell layer tumour model." British journal of cancer 97, no. 2 (2007): 194-200.
    
* Drug C - Paclitaxel

    * Krouglova, Tatiana, Jo Vercammen, and Yves Engelborghs. "Correct diffusion coefficients of proteins in fluorescence correlation spectroscopy. Application to tubulin oligomers induced by Mg2+ and Paclitaxel." Biophysical journal 87, no. 4 (2004): 2635-2646.

    * Vasantha, J., G. Kannan, T. Goud, T. Palani, R. Vanitha, R. Anitha, and J. M. M. Priya. "Pharmacokinetic evaluation of Paclitaxel in South Indian cancer patients: a prospective study." Journal of Young Pharmacists 3, no. 4 (2011): 322-328.


```python
data = np.loadtxt(open("data/cancer_spheroid.csv", "rb"), delimiter=",", skiprows=0)
r_a = data[:,0]
d_a = data[:,1]
r_b = data[:,2]
d_b = data[:,3]
r_c = data[:,4]
d_c = data[:,5]
```

```python
diff_a = 6.7e-11
diff_b = 1.78e-11
diff_c = 4.1e-11
k_a = 5.8e-5
k_b = 2.7e-3
k_c = 9.33e-4
```

Plot the data for the three drugs as penetration depth versus radius of tumor (i.e. with dimensions)

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

plt.scatter(r_a * 1e6, d_a * 1e6)
plt.scatter(r_b * 1e6, d_b * 1e6)
plt.scatter(r_c * 1e6, d_c * 1e6)

plt.yscale('log')

plt.xlabel('Radius of tumor spheroid ($\mu m$)')
plt.ylabel('Penetration depth ($\mu m$)')

plt.show()
```

Calculate the Thiele modulus and the radius saturation fraction for the three drugs

```python
thiele_a = np.sqrt(k_a * r_a ** 2 / diff_a)
thiele_b = np.sqrt(k_b * r_b ** 2 / diff_b)
thiele_c = np.sqrt(k_c * r_c ** 2 / diff_c)
delta_a = d_a / r_a
delta_b = d_b / r_b
delta_c = d_c / r_c
```

Plot the data using dimensionless groups

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

plt.scatter(thiele_a, delta_a)
plt.scatter(thiele_b, delta_b)
plt.scatter(thiele_c, delta_c)

plt.yscale('log')
plt.xscale('log')

plt.ylabel('Radius saturation fraction $\delta$')
plt.xlabel('Thiele modulus $\Phi$')

plt.show()
```
