Logistic Regression for Modeling Binary Outcomes
========================================

In this Jupyter Notebook...

* Perform a logistic regression to determine COVID survivability with age and pre-existing conditions

* Impliment a Newton-Raphson optimization of the fit variables

* Show the 10%, 50%, and 90% survivability bands versus age and number of pre-existing conditions

Import the necessary modules

```python
import numpy as np
import sys
sys.path.append('../')
#import chemeng177.logistic_regression as logr
import matplotlib.pyplot as plt
```

Load the data [anonymised COVID patient-related information from Mexican government https://www.gob.mx/salud/documentos/datos-abiertos-152127]

```python
data = np.loadtxt(open("data/covid_mexico.csv", "rb"), delimiter=",", skiprows=1)

num_exp = 10000

survive = np.array([])
age = np.array([])
pre_exist = np.array([])
for i in range(num_exp):
    if data[i,5] < 100:
        survive = np.append(survive,data[i,0])
        age = np.append(age,data[i,5])
        pre_exist_i = np.sum(2-data[i,7:17])
        if pre_exist_i < 0:
            pre_exist_i = 0
        pre_exist = np.append(pre_exist,pre_exist_i)
```

Define auxialiary functions for the analysis

```python
def logit(t, a):
    r"""
    Logit function for logistic regression

    Parameters
    ----------
    t : float array
        Covariate vector
    a : float array
        Fit parameter vector

    Returns
    -------
    g_inv : float
        Value of the logit function

    """
    t_dot_a = np.dot(t, a)
    g_inv = np.exp(t_dot_a) / (1 + np.exp(t_dot_a))

    return g_inv


def l_gradient(x, t, a):
    r"""
    Calculate the gradient of the log-likelihood with respect to the covariate vector

    Parameters
    ----------
    x : int array
        Vector of the discrete dependent variable (zeros and ones)
    t : float array
        Covariate vector
    a : float array
        Fit parameter vector

    Returns
    -------
    l_grad : float array
        Gradient of the log-likelihood

    """

    g_inv = logit(t, a)
    l_grad = np.dot(np.transpose(t), x - g_inv)

    return l_grad


def l_hessian(x, t, a):
    r"""
    Calculate the Hessian of the log-likelihood with respect to the covariate vector

    Parameters
    ----------
    x : int array
        Vector of the discrete dependent variable (zeros and ones)
    t : float array
        Covariate vector
    a : float array
        Fit parameter vector

    Returns
    -------
    l_hess : float array
        Hessian of the log-likelihood

    """

    g_inv = logit(t, a)
    l_hess = np.zeros((3, 3))
    l_hess[0, :] = - np.dot(np.transpose(t), t[:, 0] * g_inv * (1 - g_inv))
    l_hess[1, :] = - np.dot(np.transpose(t), t[:, 1] * g_inv * (1 - g_inv))
    l_hess[2, :] = - np.dot(np.transpose(t), t[:, 2] * g_inv * (1 - g_inv))

    return l_hess
```

Visualize the data

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

for i in range(num_exp):
    if survive[i] == 1:
        plt.scatter(age[i], pre_exist[i],color = 'C0')
    else:
        plt.scatter(age[i], pre_exist[i],color = 'C1')

plt.xlabel('Age')
plt.ylabel('Pre-existing conditions')

plt.show()
```

Plot the histogram of the ages

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

plt.hist(age, bins=20)

plt.xlabel('Age')
plt.ylabel('Survival')

plt.show()
```

Perform a logistic regression of the data of survival versus covariates age and pre-existing conditions

```python
num_opt = 100

t = np.transpose(np.array([np.ones(len(age)), age, pre_exist]))
x = survive
a = np.array([1e-3,2e-3,3e-3]) # Initial guess for the model parameters

for i_opt in range(num_opt):
    l_grad = l_gradient(x, t, a)
    l_hess = l_hessian(survive, t, a)
    a = a - np.dot(np.linalg.inv(l_hess), l_grad)
```

```python
plt.figure()
font = {'family' : 'serif',
        'weight':'normal',
        'size': 18}
plt.rc('font', **font)

for i in range(num_exp):
    if survive[i] == 1:
        plt.scatter(age[i], pre_exist[i],color = 'C0')
    else:
        plt.scatter(age[i], pre_exist[i],color = 'C1')

        
age_vec = np.linspace(0, 100, 100)

alpha = 0.1
plt.plot(age_vec, - a[1] / a[2] * age_vec - a[0] / a[2] + np.log(alpha / (1 - alpha)) / a[2], 'k:')
alpha = 0.5
plt.plot(age_vec, - a[1] / a[2] * age_vec - a[0] / a[2] + np.log(alpha / (1 - alpha)) / a[2], 'k-')        
alpha = 0.9
plt.plot(age_vec, - a[1] / a[2] * age_vec - a[0] / a[2] + np.log(alpha / (1 - alpha)) / a[2], 'k--')        


plt.ylim([-1, 8])
plt.xlim([0, 100])

plt.xlabel('Age')
plt.ylabel('Number of pre-existing conditions')

plt.savefig('figures/logistic_regression/covid.pdf', bbox_inches='tight')
plt.show()
```
